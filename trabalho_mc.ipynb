```python
import pandas as pd

# Certifique-se de que o DataFrame 'df' está carregado corretamente
# e que ele contém as colunas 'ano', 'id_municipio_nome', 'estado' e 'desmatado'.
# Se necessário, execute novamente o bloco de código que recarrega 'df'
# com a 'query_df_com_estado' (a que inclui a coluna 'estado').

print("--- Município e Estado com Maior Área Desmatada por Ano ---")

# Obtém todos os anos únicos presentes no DataFrame e os ordena
anos_unicos = sorted(df['ano'].unique())

# Lista para armazenar os resultados de cada ano
resultados_por_ano = []

for ano in anos_unicos:
    # Filtra o DataFrame para o ano atual
    df_do_ano = df[df['ano'] == ano]

    if not df_do_ano.empty:
        # Agrupa por município e estado, e soma o desmatamento para esse ano
        desmatamento_municipio_estado = df_do_ano.groupby(['id_municipio_nome', 'estado'])['desmatado'].sum().reset_index()

        # Encontra a linha com o maior valor de 'desmatado' para o ano atual
        if not desmatamento_municipio_estado.empty:
            municipio_mais_desmatado_no_ano = desmatamento_municipio_estado.loc[desmatamento_municipio_estado['desmatado'].idxmax()]

            resultados_por_ano.append({
                'Ano': ano,
                'Município': municipio_mais_desmatado_no_ano['id_municipio_nome'],
                'Estado': municipio_mais_desmatado_no_ano['estado'],
                'Área Desmatada (km²)': municipio_mais_desmatado_no_ano['desmatado']
            })
        else:
            # Caso não haja desmatamento registrado para nenhum município nesse ano (o que é raro, mas possível)
            resultados_por_ano.append({
                'Ano': ano,
                'Município': 'N/A',
                'Estado': 'N/A',
                'Área Desmatada (km²)': 0
            })
    else:
        # Caso o ano não exista no DataFrame
        resultados_por_ano.append({
            'Ano': ano,
            'Município': 'N/A',
            'Estado': 'N/A',
            'Área Desmatada (km²)': 0
        })

# Converte a lista de dicionários em um DataFrame para uma exibição tabular
tabela_anual_maior_desmatamento = pd.DataFrame(resultados_por_ano)

print(tabela_anual_maior_desmatamento)
```

```python
import plotly.express as px
import pandas as pd

# Certifique-se de que o DataFrame 'df' está carregado corretamente
# com as colunas 'ano', 'estado' e 'desmatado'.
# Se necessário, execute o bloco de código que recarrega 'df' com a 'query_df_com_estado'
# antes de rodar este código.

# 1. Filtrar os dados para o período entre 2000 e 2023
df_periodo_2000_2023 = df[(df['ano'] >= 2000) & (df['ano'] <= 2023)]

if df_periodo_2000_2023.empty:
    print("Não há dados de desmatamento para o período entre 2000 e 2023 no DataFrame 'df'.")
else:
    # 2. Agrupar por estado e somar a área desmatada no período
    desmatamento_por_estado_total = df_periodo_2000_2023.groupby('estado')['desmatado'].sum().reset_index()
    desmatamento_por_estado_total.columns = ['Estado', 'Área Desmatada (km²)']

    # Opcional: Ordenar os estados pela área desmatada para que o gráfico fique mais claro
    desmatamento_por_estado_total = desmatamento_por_estado_total.sort_values(by='Área Desmatada (km²)', ascending=True)

    # 3. Criar o gráfico de barras horizontais para Estados
    fig_estado_barras = px.bar(desmatamento_por_estado_total,
                               x='Área Desmatada (km²)',
                               y='Estado',
                               title='Área Desmatada por Estado (2000-2023)',
                               orientation='h', # Define as barras como horizontais
                               text='Área Desmatada (km²)', # Exibe o valor da área desmatada na barra
                               height=600) # Ajusta a altura do gráfico para melhor visualização

    # Ajustar o formato do texto para ter duas casas decimais e posicioná-lo fora da barra
    fig_estado_barras.update_traces(texttemplate='%{text:.2f}', textposition='outside')
    fig_estado_barras.update_layout(uniformtext_minsize=8, uniformtext_mode='hide') # Esconder texto se for muito pequeno

    fig_estado_barras.show()
```

Agora, coletamos os dados sobre o efetivo dos rebanhos do IBGE PPM, também através da biblioteca `basedosdados`. A query busca informações sobre o ano, estado, município, tipo de rebanho e quantidade.

```python
import basedosdados as bd
import pandas as pd

# Defina o seu billing_project_id
billing_id = "trabalho-mc"

# Query SQL para obter dados de efetivo de rebanhos
query_rebanhos = """
  SELECT
    dados.ano as ano,
    dados.sigla_uf AS sigla_uf,
    diretorio_sigla_uf.nome AS sigla_uf_nome,
    dados.id_municipio AS id_municipio,
    diretorio_id_municipio.nome AS id_municipio_nome,
    dados.tipo_rebanho as tipo_rebanho,
    dados.quantidade as quantidade
FROM `basedosdados.br_ibge_ppm.efetivo_rebanhos` AS dados
LEFT JOIN (SELECT DISTINCT sigla,nome  FROM `basedosdados.br_bd_diretorios_brasil.uf`) AS diretorio_sigla_uf
    ON dados.sigla_uf = diretorio_sigla_uf.sigla
LEFT JOIN (SELECT DISTINCT id_municipio,nome  FROM `basedosdados.br_bd_diretorios_brasil.municipio`) AS diretorio_id_municipio
    ON dados.id_municipio = diretorio_id_municipio.id_municipio
"""

print("Executando a consulta de dados de rebanhos... Isso pode levar um tempo.")
df_rebanhos = bd.read_sql(query_rebanhos, billing_project_id=billing_id)
print("Consulta de rebanhos concluída e dados carregados no DataFrame 'df_rebanhos'.")

# Exibir informações e primeiras linhas
print("\nInformações gerais do DataFrame de rebanhos:")
df_rebanhos.info()
print("\nPrimeiras 5 linhas do DataFrame de rebanhos:")
display(df_rebanhos.head())

# Salvar os dados de rebanhos em um CSV
df_rebanhos.to_csv("rebanhos_efetivos.csv", index=False)
print("\nDados de rebanhos salvos em 'rebanhos_efetivos.csv'")
```

### 5. Cruzamento dos Dados

Combinamos os DataFrames de desmatamento e rebanhos em um único DataFrame `df_cruzado` utilizando as colunas `ano` e `id_municipio` como chaves de junção. Isso nos permite analisar a relação entre as duas variáveis no mesmo nível geográfico e temporal.

```python
import pandas as pd

# Carregar os DataFrames se não estiverem carregados na sessão atual
try:
    df_prodes = pd.read_csv("dados_prodes_municipio_bioma.csv")
    print("DataFrame 'df_prodes' carregado.")
except FileNotFoundError:
    print("Erro: 'dados_prodes_municipio_bioma.csv' não encontrado. Execute a célula de coleta de dados de desmatamento.")
    # Certifique-se de que a célula anterior foi executada com sucesso
    df_prodes = None

try:
    df_rebanhos = pd.read_csv("rebanhos_efetivos.csv")
    print("DataFrame 'df_rebanhos' carregado.")
except FileNotFoundError:
     print("Erro: 'rebanhos_efetivos.csv' não encontrado. Execute a célula de coleta de dados de rebanhos.")
     df_rebanhos = None


if df_prodes is not None and df_rebanhos is not None:
    print("\nRealizando o cruzamento dos DataFrames...")

    # Garantir que a coluna id_municipio tenha o mesmo formato em ambos os DataFrames (string)
    df_prodes['id_municipio'] = df_prodes['id_municipio'].astype(str)
    df_rebanhos['id_municipio'] = df_rebanhos['id_municipio'].astype(str)

    # Realizar a junção (merge) dos DataFrames
    # Usamos 'inner' merge para incluir apenas os municípios e anos presentes em ambos os datasets
    df_cruzado = pd.merge(
        df_rebanhos,
        df_prodes,
        left_on=['ano', 'id_municipio'],
        right_on=['ano', 'id_municipio'],
        how='inner',
        suffixes=('_rebanho', '_prodes') # Sufixos para colunas com nomes duplicados
    )

    print("Cruzamento concluído.")
    print("\nDimensões do DataFrame cruzado:", df_cruzado.shape)
    print("\nPrimeiras 5 linhas do DataFrame cruzado:")
    display(df_cruzado.head())
    print("\nColunas do DataFrame cruzado:")
    print(df_cruzado.columns)

    # Renomear colunas relevantes no DataFrame cruzado para maior clareza
    df_cruzado.rename(columns={
        'desmatado': 'desmatamento_km2',
        'quantidade': 'total_rebanho',
        'id_municipio_nome_rebanho': 'municipio_nome', # Nome do município vindo de rebanhos
        'sigla_uf_rebanho': 'estado_sigla' # Sigla do estado vindo de rebanhos
        # Nota: id_municipio_nome_prodes e sigla_uf_prodes também existem
        # se ambas as tabelas tinham a coluna. Escolhemos uma para renomear.
    }, inplace=True)

    print("\nColunas do DataFrame cruzado após renomear:")
    print(df_cruzado.columns)
    display(df_cruzado.head()) # Mostrar head novamente com nomes novos

    # Opcional: Salvar o DataFrame cruzado
    df_cruzado.to_csv("dados_rebanhos_prodes_cruzado.csv", index=False)
    print(f"\nDados cruzados salvos em 'dados_rebanhos_prodes_cruzado.csv'")

else:
    print("\nNão foi possível realizar o cruzamento. Certifique-se de que ambos os DataFrames ('df_prodes', 'df_rebanhos') foram carregados corretamente.")
```

### 6. Análise de Desmatamento

Realizamos análises focadas nos dados de desmatamento:

*   **Total de Desmatamento por Bioma ao Longo dos Anos:** Agrupamos os dados de desmatamento por bioma e ano para ver a evolução.
*   **Total de Desmatamento por Bioma (Total):** Somamos o desmatamento total para cada bioma ao longo de todos os anos disponíveis.
*   **Top 10 Municípios com Maior Desmatamento (em anos específicos e total):** Identificamos os municípios com maior área desmatada em 2020, 2022 e no período total.
*   **Município e Estado com Maior Desmatamento por Ano:** Encontramos o município com a maior área desmatada para cada ano individualmente.
*   **Visualização da Participação por Bioma:** Criamos um gráfico de pizza para mostrar a proporção do desmatamento total por bioma.
*   **Visualização do Desmatamento Total por Ano:** Criamos uma tabela e um gráfico de barras para mostrar a evolução anual do desmatamento.
*   **Visualização do Desmatamento por Estado:** Criamos um gráfico de barras horizontais para mostrar o desmatamento total por estado.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px # Para gráfico de barras interativo

# Carregar o DataFrame de desmatamento se não estiver carregado
try:
    df_prodes = pd.read_csv("dados_prodes_municipio_bioma.csv")
    print("DataFrame 'df_prodes' carregado.")
except FileNotFoundError:
    print("Erro: 'dados_prodes_municipio_bioma.csv' não encontrado. Execute a célula de coleta de dados de desmatamento.")
    df_prodes = None

if df_prodes is not None:
    print("\n--- Análise de Desmatamento ---")

    # Total de Desmatamento por Bioma ao Longo dos Anos
    desmatamento_por_bioma_ano = df_prodes.groupby(['bioma', 'ano'])['desmatado'].sum().reset_index()
    print("\nTotal de desmatamento por bioma ao longo dos anos (Primeiras 5 linhas):")
    display(desmatamento_por_bioma_ano.head())

    # Total de Desmatamento por Bioma (Total)
    desmatamento_total_por_bioma = df_prodes.groupby('bioma')['desmatado'].sum().reset_index()
    desmatamento_total_por_bioma = desmatamento_total_por_bioma.sort_values(by='desmatado', ascending=False)
    print("\nDesmatamento total por bioma (todos os anos combinados):")
    display(desmatamento_total_por_bioma)

    # Visualização da Participação por Bioma (Gráfico de Pizza)
    plt.figure(figsize=(10, 10))
    plt.pie(
        desmatamento_total_por_bioma['desmatado'],
        labels=desmatamento_total_por_bioma['bioma'],
        autopct='%1.1f%%',
        startangle=90,
        pctdistance=0.85,
        wedgeprops={'edgecolor': 'black'}
    )
    plt.title('Participação de Cada Bioma no Desmatamento Total (Todos os Anos)')
    plt.axis('equal')
    plt.show()

    # Top 10 Municípios com Maior Desmatamento em anos específicos
    for ano_analise in [2020, 2022]:
        df_ano = df_prodes[df_prodes['ano'] == ano_analise]
        if not df_ano.empty:
            top_10_municipios_ano = df_ano.groupby(['id_municipio_nome', 'estado'])['desmatado'].sum().nlargest(10).reset_index()
            top_10_municipios_ano.columns = ['Município', 'Estado', 'Área Desmatada (km²)']
            print(f"\n--- Top 10 Municípios com Maior Área Desmatada em {ano_analise} (com Estado) ---")
            display(top_10_municipios_ano)
        else:
            print(f"\nNão há dados de desmatamento para o ano de {ano_analise} no DataFrame.")

    # Top 10 Municípios com Maior Desmatamento Total
    top_10_municipios_total_estado = df_prodes.groupby(['id_municipio_nome', 'estado'])['desmatado'].sum().nlargest(10).reset_index()
    top_10_municipios_total_estado.columns = ['Município', 'Estado', 'Área Desmatada Total (km²)']
    print("\n--- Top 10 Municípios com Maior Área Desmatada Total (Todos os Anos, com Estado) ---")
    display(top_10_municipios_total_estado)

    # Município e Estado com Maior Área Desmatada por Ano
    print("\n--- Município e Estado com Maior Área Desmatada por Ano ---")
    anos_unicos = sorted(df_prodes['ano'].unique())
    resultados_por_ano = []
    for ano in anos_unicos:
        df_do_ano = df_prodes[df_prodes['ano'] == ano]
        if not df_do_ano.empty:
            desmatamento_municipio_estado = df_do_ano.groupby(['id_municipio_nome', 'estado'])['desmatado'].sum().reset_index()
            if not desmatamento_municipio_estado.empty:
                municipio_mais_desmatado_no_ano = desmatamento_municipio_estado.loc[desmatamento_municipio_estado['desmatado'].idxmax()]
                resultados_por_ano.append({
                    'Ano': ano,
                    'Município': municipio_mais_desmatado_no_ano['id_municipio_nome'],
                    'Estado': municipio_mais_desmatado_no_ano['estado'],
                    'Área Desmatada (km²)': municipio_mais_desmatado_no_ano['desmatado']
                })
            else:
                resultados_por_ano.append({'Ano': ano, 'Município': 'N/A', 'Estado': 'N/A', 'Área Desmatada (km²)': 0})
        else:
             resultados_por_ano.append({'Ano': ano, 'Município': 'N/A', 'Estado': 'N/A', 'Área Desmatada (km²)': 0})
    tabela_anual_maior_desmatamento = pd.DataFrame(resultados_por_ano)
    display(tabela_anual_maior_desmatamento)

    # Visualização do Desmatamento Total por Ano (Tabela e Gráfico)
    desmatamento_por_ano_tabela = df_prodes.groupby('ano')['desmatado'].sum().reset_index()
    desmatamento_por_ano_tabela.columns = ['Ano', 'Área Desmatada (km²)']
    print("\n--- Área Desmatada Total por Ano ---")
    display(desmatamento_por_ano_tabela)

    plt.figure(figsize=(12, 6))
    sns.lineplot(data=desmatamento_por_ano_tabela, x='Ano', y='Área Desmatada (km²)')
    plt.title('Evolução da Área Desmatada Total por Ano (2000-2023)')
    plt.xlabel('Ano')
    plt.ylabel('Área Desmatada (km²)')
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.show()


    # Visualização do Desmatamento por Estado (Gráfico de Barras Horizontais)
    desmatamento_por_estado_total = df_prodes.groupby('estado')['desmatado'].sum().reset_index()
    desmatamento_por_estado_total.columns = ['Estado', 'Área Desmatada (km²)']
    desmatamento_por_estado_total = desmatamento_por_estado_total.sort_values(by='Área Desmatada (km²)', ascending=True)

    fig_estado_barras = px.bar(desmatamento_por_estado_total,
                               x='Área Desmatada (km²)',
                               y='Estado',
                               title='Área Desmatada por Estado (2000-2023)',
                               orientation='h',
                               text='Área Desmatada (km²)',
                               height=800) # Ajusta a altura do gráfico

    fig_estado_barras.update_traces(texttemplate='%{text:.2f}', textposition='outside')
    fig_estado_barras.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')

    fig_estado_barras.show()

else:
    print("\nAnálise de Desmatamento não executada pois o DataFrame 'df_prodes' não foi carregado.")
```

### 7. Análise da Relação Desmatamento vs. Rebanho Bovino

Nesta etapa, investigamos a correlação entre o desmatamento e o efetivo de rebanho bovino, focando em dados agregados em nível nacional por ano.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np # Importar numpy para np.log1p se for usar no mapa

# Carregar o DataFrame cruzado se não estiver carregado
try:
    df_cruzado = pd.read_csv("dados_rebanhos_prodes_cruzado.csv")
    print("DataFrame 'df_cruzado' carregado.")
    # Renomear colunas caso o CSV tenha sido carregado diretamente e não após o merge na célula 5
    df_cruzado.rename(columns={
        'desmatado': 'desmatamento_km2',
        'quantidade': 'total_rebanho',
        'id_municipio_nome_rebanho': 'municipio_nome',
        'sigla_uf_rebanho': 'estado_sigla'
    }, inplace=True)
    # Filtrar apenas para Bovinos, se necessário
    df_cruzado_bovinos = df_cruzado[df_cruzado['tipo_rebanho'] == 'Bovino'].copy()
    print("Filtrado para tipo_rebanho = 'Bovino'.")

except FileNotFoundError:
    print("Erro: 'dados_rebanhos_prodes_cruzado.csv' não encontrado. Execute a célula de cruzamento de dados (Passo 5).")
    df_cruzado_bovinos = None

if df_cruzado_bovinos is not None:
    print("\n--- Análise da Relação Desmatamento vs. Rebanho Bovino (Nacional Anual) ---")

    # Agregando os dados para o nível nacional anual
    df_nacional_anual = df_cruzado_bovinos.groupby('ano').agg(
        desmatamento_km2=('desmatamento_km2', 'sum'),
        total_bovinos=('total_rebanho', 'sum') # Usar total_rebanho após renomear
    ).reset_index()

    print("Dados Nacionais Anuais Agregados para o Gráfico:")
    display(df_nacional_anual.head())

    # --- Criar o Gráfico de Dispersão com Linhas ---
    plt.figure(figsize=(10, 6))

    # Plotar os pontos de dispersão (cada ponto representa um ano)
    plt.scatter(df_nacional_anual['desmatamento_km2'], df_nacional_anual['total_bovinos'],
                color='blue', alpha=0.7, s=80, label='Dados Anuais')

    # Adicionar linhas conectando os pontos na ordem cronológica (pelo ano)
    df_nacional_anual_sorted = df_nacional_anual.sort_values(by='ano')
    plt.plot(df_nacional_anual_sorted['desmatamento_km2'], df_nacional_anual_sorted['total_bovinos'],
             color='red', linestyle='--', linewidth=1.5, alpha=0.7, label='Progresso Temporal')

    # Adicionar rótulos e título
    plt.title('Correlação entre Desmatamento (PRODES) e Efetivo de Bovinos (PPM) no Brasil (Nacional Anual)', fontsize=14)
    plt.xlabel('Desmatamento Anual (km²)', fontsize=12)
    plt.ylabel('Efetivo de Bovinos (cabeças)', fontsize=12)

    # Anotar os anos em cada ponto
    for i, row in df_nacional_anual_sorted.iterrows():
        plt.annotate(f"{int(row['ano'])}", (row['desmatamento_km2'], row['total_bovinos']),
                     textcoords="offset points", xytext=(5,5), ha='center', fontsize=9, color='darkgreen')

    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # --- Calcular e exibir o coeficiente de correlação de Pearson ---
    correlation_coefficient = df_nacional_anual['desmatamento_km2'].corr(df_nacional_anual['total_bovinos'])
    print(f"\n**Coeficiente de Correlação de Pearson (Desmatamento x Bovinos, Nacional Anual): {correlation_coefficient:.2f}**")

    if correlation_coefficient > 0.7:
        print("Isso sugere uma correlação positiva forte: à medida que o desmatamento aumenta, o efetivo de bovinos também tende a aumentar.")
    elif correlation_coefficient < -0.7:
        print("Isso sugere uma correlação negativa forte.")
    elif 0.3 <= correlation_coefficient <= 0.7:
        print("Isso sugere uma correlação positiva moderada.")
    elif -0.7 <= correlation_coefficient <= -0.3:
        print("Isso sugere uma correlação negativa moderada.")
    else:
        print("Isso sugere uma correlação linear fraca ou inexistente.")

else:
     print("\nAnálise da Relação não executada pois o DataFrame 'df_cruzado_bovinos' não foi carregado ou filtrado corretamente.")
```
